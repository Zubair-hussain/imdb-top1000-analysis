# **Dataset Description (Purpose/Context)**

  The **IMDb Movie** Dataset contains detailed information about the top 1,000 movies listed on IMDb, including data points such as genre, release year, director, runtime, IMDb rating, Metascore, box office revenue, and vote count.

  This dataset is ideal for performing exploratory data analysis (EDA) in the entertainment domain. It enables us to uncover trends in movie performance, identify what genres are most popular, and determine factors that might influence critical acclaim or commercial success.

  The analysis can help studios, content platforms, and filmmakers make data-driven decisions about what types of films resonate most with audiences.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/peetck/IMDB-Top1000-Movies/master/IMDB-Movie-Data.csv"
df = pd.read_csv(url)
df.head()
```

# Rename Columns for **clarity**

```python
df.columns = ['Rank', 'Title', 'Genre', 'Description', 'Director', 'Actors',
              'Year', 'Runtime', 'Rating', 'Votes', 'Revenue', 'Metascore']

df
```

# Check for mi**ssing data**

```python
print("Missing values:\n", df.isnull().sum())
```


#Drop duplicates '(if any)'

```python
df.drop_duplicates(inplace=True)
df
```

#Convert data types

```python
df['Year'] = df['Year'].astype(int)
df['Runtime'] = df['Runtime'].astype(int)
df['Votes'] = df['Votes'].astype(int)
df['Metascore'] = df['Metascore'].astype(int)
df['Revenue'] = df['Revenue'].astype(float)

df.dtypes
```

# Confirm Data is Clean

```python
# Confirm data is clean
print("\nCleaned Data Summary:")
print(df.info())
```

# Exploratory Data Analysis (EDA) & Visualization

Set default styles

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

# Histogram of IMDb Ratings

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of IMDb Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Bar Chart: Top 10 Most Common Genres

```python
# Explode genres into individual rows
genre_split = df['Genre'].str.split(',', expand=False)
genres_exploded = df.assign(Genre=genre_split).explode('Genre')
top_genres = genres_exploded['Genre'].value_counts().head(10)

top_genres.plot(kind='bar', color='salmon')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

# Line Chart: Average Revenue by Year

```python
revenue_by_year = df.groupby('Year')['Revenue'].mean()
revenue_by_year.plot(kind='line', marker='o', color='green')
plt.title('Average Revenue by Year')
plt.ylabel('Revenue (Millions)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

# Heatmap: Correlation Between Numerical Features

```python
correlation = df[['Runtime', 'Rating', 'Votes', 'Revenue', 'Metascore']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Scatter Plot: Revenue vs IMDb Rating

```python
sns.scatterplot(data=df, x='Rating', y='Revenue', hue='Genre', legend=False, palette='husl')
plt.title('Revenue vs IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Revenue (Millions)')
plt.show()
```
